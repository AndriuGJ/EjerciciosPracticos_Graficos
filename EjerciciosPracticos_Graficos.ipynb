Agrupa los siguientes datos en intervalos de amplitud 2:
2, 3, 5, 1, 6, 7, 8, 0, 8, 9, 6, 4, 4, 3, 2, 6, 3, 3, 2, 1, 1, 5, 6, 7, 4, 3, 2, 5, 4, 6
Construye una tabla de frecuencias completa y dibuja el histograma
correspondiente.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

datos = np.array([2, 3, 5, 1, 6, 7, 8, 0, 8, 9, 6, 4, 4, 3, 2, 6, 3, 3, 2, 1, 1, 5, 6, 7, 4, 3, 2, 5, 4, 6])

valor_minimo = datos.min()
valor_maximo = datos.max()
ancho_intervalo = 2
intervalos = np.arange(valor_minimo, valor_maximo + ancho_intervalo, ancho_intervalo)

print("Intervalos (bins):", intervalos)

tabla_frecuencia = pd.cut(datos, bins=intervalos, right=False, include_lowest=True).value_counts().sort_index()

print("\nRecuentos de frecuencia para cada intervalo:")
display(tabla_frecuencia)

frecuencia_relativa = tabla_frecuencia / tabla_frecuencia.sum()

print("\nFrecuencias relativas:")
display(frecuencia_relativa)

frecuencia_acumulada = tabla_frecuencia.cumsum()
frecuencia_relativa_acumulada = frecuencia_relativa.cumsum()

print("\nFrecuencias acumuladas:")
display(frecuencia_acumulada)
print("\nFrecuencias relativas acumuladas:")
display(frecuencia_relativa_acumulada)

df_frecuencia = pd.DataFrame({
    'Frecuencia': tabla_frecuencia,
    'Frecuencia Relativa': frecuencia_relativa,
    'Frecuencia Acumulada': frecuencia_acumulada,
    'Frecuencia Relativa Acumulada': frecuencia_relativa_acumulada
})

print("\nTabla de frecuencias completa:")
display(df_frecuencia)

plt.figure(figsize=(10, 6))
plt.hist(datos, bins=intervalos, edgecolor='black', alpha=0.7)
plt.xticks(intervalos)
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.title("Histograma de datos con intervalos de amplitud 2")
plt.grid(axis='y', alpha=0.75)
plt.show()
```

La distribucin del tiempo libre semanal de un grupo de personas es la
siguiente:
o Deporte: 30%
o Lectura: 20%
o Cine: 15%
o Viajes: 10%
o Otras actividades: 25%
a) Representa estos datos mediante un diagrama de sectores.
b) Calcula el ngulo correspondiente a cada sector.


```python
tiempo_libre = ['Deporte', 'Lectura', 'Cine', 'Viajes', 'Otras actividades']
porcentajes = [30, 20, 15, 10, 25]

fig, ax = plt.subplots()
ax.pie(porcentajes, labels=tiempo_libre, autopct='%1.1f%%')
ax.set_title('Distribucin del tiempo libre semanal')
plt.show()

#Angulo de cada sector en grados
angulos_grados = [p * 3.6 for p in porcentajes]
print("ngulos correspondientes a cada sector:")
for sector, angulo in zip(tiempo_libre, angulos_grados):
    print(f"{sector}: {angulo:.2f} grados")
```
